```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv('Batsman_All - Copy.csv')
df.head()
```

```python
df.describe()
```

```python
df['rating'] = df['Runs'] / df['Inns'] + df['Runs'] / df['BF'] + df['hundred'] + df['fifty'] / df['Inns'] + df['four']*4 / df['Inns']
```

```python
df.head()
```

```python
df['res'] = df['rating'].apply(lambda x: '1' if x >= 20 else '0')
```

```python
df.head()
```

```python
target=df.res
inputs=df.drop(['res','Player','Span','rating'],axis='columns')
```

```python
X_train, X_test, y_train, y_test= train_test_split(inputs,target,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_test[:5]
```

```python
m=model.predict(X_test)
```

```python
model.predict_proba(X_test[:10])
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, m))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(RandomForestClassifier(n_estimators=100),X_train, y_train, cv=5)
```
